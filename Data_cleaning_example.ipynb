```python
df["Peak"]=pd.to_numeric(df["Peak"],errors="coerce")
df["All Time Peak"] = df["All Time Peak"].astype(str).str.replace(r'\[.*?\]', '', regex=True)
df["All Time Peak"] = pd.to_numeric(df["All Time Peak"], errors="coerce")
df["All Time Peak"].fillna(df["All Time Peak"].mean(), inplace=True)
```

```python
df["Peak"].fillna(np.mean(df["Peak"]))
```
